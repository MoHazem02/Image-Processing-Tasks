# Requirements ðŸ—’

### 1. Generate a random matrix \( M \times N \) of values 0-255
### 2. Generate a correct form of Bayer Filter
### 3. Map the correct values and interpolate the missing ones
### 4. Transform to HSV color space
### 5. Show all the inputs and outputs




```python
import numpy as np
import matplotlib.pyplot as plt

M = 8
N = 8

# Req 1: Ranodm matrix (MxN) with values between 0 and 255 âœ…
RANDOM_MATRIX = np.random.randint(0, 256, (M, N, 3), dtype=np.uint8) 

# Display RANDOM MATRIX
plt.imshow(RANDOM_MATRIX)
plt.title('Random Matrix')
plt.axis('off')
plt.show()
```

```python
# Req 2: Generate Bayer Filter âœ…
BAYER_FILTER = np.zeros((M, N, 3), dtype=np.uint8)

# Red channel
BAYER_FILTER[0::2, 0::2, 0] = 255
# Green channel
BAYER_FILTER[0::2, 1::2, 1] = 255
BAYER_FILTER[1::2, 0::2, 1] = 255
# Blue channel
BAYER_FILTER[1::2, 1::2, 2] = 255

# Display Bayer Filter
plt.imshow(BAYER_FILTER)
plt.title('Bayer Filter')
plt.axis('off')
plt.show()
```

```python
# Req 3: Apply Bayer Filter to Random Matrix âœ…
FILTERED_MATRIX = BAYER_FILTER * RANDOM_MATRIX

# Display Filtered Matrix
plt.imshow(FILTERED_MATRIX)
plt.title('Filtered Matrix')
plt.axis('off')
plt.show()
```

```python
import cv2
# Req 3.2: Perform Bilinear Interpolation using OpenCV âœ…
INTERPOLATED_IMAGE = cv2.resize(FILTERED_MATRIX, (M, N), interpolation=cv2.INTER_LINEAR)

# Display Interpolated Image
plt.imshow(INTERPOLATED_IMAGE)
plt.title('Interpolated Image')
plt.axis('off')
```

```python
# Req 4: Transform to HSV color space âœ…
HSV_IMAGE = cv2.cvtColor(INTERPOLATED_IMAGE, cv2.COLOR_RGB2HSV)

# Display HSV Image
plt.imshow(HSV_IMAGE)
plt.title('HSV Image')
plt.axis('off')
plt.show()
```
