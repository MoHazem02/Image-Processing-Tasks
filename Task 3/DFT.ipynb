# Discrete Fourier Transform (DFT) on Two Images
This notebook demonstrates the application of DFT on two images and interprets the resulting magnitude and phase images.

```python
# Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the first image
image1 = cv2.imread('monkey.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(image1, cmap='gray')
plt.title('Image 1 - Grayscale')
plt.axis('off')
plt.show()

# Load the second image
image2 = cv2.imread('Pic1.webp', cv2.IMREAD_GRAYSCALE)
plt.imshow(image2, cmap='gray')
plt.title('Image 2 - Grayscale')
plt.axis('off')
plt.show()
```

```python
# Perform DFT on the first image
dft1 = cv2.dft(np.float32(image1), flags=cv2.DFT_COMPLEX_OUTPUT)
dft1_shift = np.fft.fftshift(dft1)
magnitude1 = 20 * np.log(cv2.magnitude(dft1_shift[:, :, 0], dft1_shift[:, :, 1]))
phase1 = np.angle(dft1_shift[:, :, 0] + 1j * dft1_shift[:, :, 1])

# Display magnitude and phase images for the first image
plt.imshow(magnitude1, cmap='gray')
plt.title('Image 1 - Magnitude Spectrum')
plt.axis('off')
plt.show()

plt.imshow(phase1, cmap='gray')
plt.title('Image 1 - Phase Spectrum')
plt.axis('off')
plt.show()
```

```python
# Perform DFT on the second image
dft2 = cv2.dft(np.float32(image2), flags=cv2.DFT_COMPLEX_OUTPUT)
dft2_shift = np.fft.fftshift(dft2)
magnitude2 = 20 * np.log(cv2.magnitude(dft2_shift[:, :, 0], dft2_shift[:, :, 1]))
phase2 = np.angle(dft2_shift[:, :, 0] + 1j * dft2_shift[:, :, 1])

# Display magnitude and phase images for the second image
plt.imshow(magnitude2, cmap='gray')
plt.title('Image 2 - Magnitude Spectrum')
plt.axis('off')
plt.show()

plt.imshow(phase2, cmap='gray')
plt.title('Image 2 - Phase Spectrum')
plt.axis('off')
plt.show()
```

## Interpretation of Image 1 Results
**Magnitude Spectrum**: Judging from the magnitude you can the the image does not have many edges and is not bright, however there are few Horizontal edges (vertical line in magnitude)
